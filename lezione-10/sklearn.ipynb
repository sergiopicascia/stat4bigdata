```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_x = np.random.normal(2, .5, 50)
a_y = np.random.normal(3, 1.2, 50)

b_x = np.random.normal(1, .3, 50)
b_y = np.random.normal(10, .8, 50)
```

```python
x = np.concatenate([a_x, b_x])
y = np.concatenate([a_y, b_y])
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2, init='random', random_state=42)
```

```python
X = np.stack([x, y], axis=1)
```

```python
model.fit(X)
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c=model.labels_)
ax.scatter(model.cluster_centers_.T[0], model.cluster_centers_.T[1])
plt.show()
```

```python
model.inertia_
```

```python
model = KMeans(n_clusters=1, init='random', random_state=42)
```

```python
model.fit(X)
```

```python
model.inertia_
```

```python
score = []

for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    score.append(model.inertia_)
```

```python
score
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(1, 11), score, marker='x')
plt.show()
```

```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, model.predict(X_test)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model, data['data'], data['target'], cv=5)
```

```python
scores.mean()
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
```

```python
scores = cross_val_score(model, data['data'], data['target'], cv=5)
```

```python
scores.mean()
```
